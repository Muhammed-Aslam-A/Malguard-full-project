```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import pickle

```

```python
data = pd.read_csv('train.csv')

data.head()
```

```python
data = data.drop(['type'], axis = 1)
data = data.drop(['android'], axis = 1)
data.dropna(axis = 1, inplace = True)

data.head()
```

```python
f=open('features.txt','w')
for col in data.columns: 
     f.write(col+'\n')
f.close()
```

```python
file1 = open('features.txt','r')
features = file1.read()
file1.close()
```

```python
features
```

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2, random_state=0)
pred_data = kmeans.fit_predict(data)

pred_data
```

```python
pred_data = list(pred_data)
data['label'] = pred_data

data.head()
column_names = data.columns.tolist()
column_names
```

```python
x = data.drop(['label'], axis = 1)
y = data['label']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.10)
x_train
```

```python
print("Number of features in training data:", x_train.shape[1])
print("Number of features in prediction data:", data.shape[1])
```

```python
clf = svm.SVC()
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
print('accuracy score: ', accuracy_score(y_test, y_pred))
```

```python
print('confusion matrix:')
print(confusion_matrix(y_test, y_pred))
```

```python
print('classification report:')
print(classification_report(y_test, y_pred))
```

```python
pickle.dump(clf, open('model.sav', 'wb'))
```
